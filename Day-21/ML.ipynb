TASK 1 — The Paradigm Shift (Traditional vs Machine Learning)

Machine Learning is a different way of solving problems compared to traditional programming. In traditional programming, the developer provides both the rules and the data, and the computer produces an output based strictly on those instructions. In contrast, in Machine Learning we provide the computer with data and expected outputs, and the system automatically discovers the underlying patterns and rules by itself. Instead of explicitly coding every condition, the model learns behavior from examples, which makes it useful for complex problems like image recognition, recommendation systems, and speech understanding where writing manual rules is impractical.

TASK 2 — The Three Branches of ML

Supervised Learning

Supervised learning is a type of machine learning where the dataset contains input features along with correct output labels. The model learns a mapping between input and output so it can predict results for new unseen data.

Unsupervised Learning

Unsupervised learning works with unlabeled data where the model tries to discover hidden patterns, similarities, or groupings within the dataset without knowing the correct answer beforehand.

Reinforcement Learning

Reinforcement learning is a learning method where an agent interacts with an environment and learns optimal behavior by receiving rewards or penalties for its actions over time.

| Feature  | Supervised Learning                    | Unsupervised Learning                         | Reinforcement Learning             |
| -------- | -------------------------------------- | --------------------------------------------- | ---------------------------------- |
| Data     | Labeled data                           | Unlabeled data                                | Interaction-based data             |
| Goal     | Predict output                         | Find hidden patterns                          | Maximize reward                    |
| Output   | Classification / Regression            | Clusters / Groups                             | Optimal actions                    |
| Examples | Spam detection, House price prediction | Customer segmentation, Market basket analysis | Self-driving cars, Game playing AI |
| Feedback | Direct correct answer                  | No answer provided                            | Reward / penalty                   |
